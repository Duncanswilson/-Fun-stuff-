\begin{align}
C(2) &= 1\\
C(n) &= C\left(\frac{n}{2}\right) + \frac{n}{2}
\end{align}
then solving the recurrance relation we see: 
\begin{align}
C(n_i) &= 2 \\
i &= lg(n)-1\\
C(n) &= C(2) + \sum_{i=1}^{lg(n)-1}\frac{n}{2^i}\\
&= 1 +(n-2)\\
&= n-1
\end{align}
